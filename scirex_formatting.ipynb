```python
from tqdm import tqdm
import json

doc_level_ann = {}
#     "paper_id" : None,
#     "body_text":[{"sentence": None, "method": None}],
#     "abstract":[{"sentence": None, "method": None}],
# }


inter_dict = {}
# section_dict ={"section":[{"sentence": "", "method": ""}]}
interd_list=[]
doc_level_list=[]
inter_text_list=[]
for item in tqdm(inter_text):
  for abstract in tqdm(item["abstract"]):
    inter_text_list = abstract["text"].split(".")
    for m_name in tqdm(method_list):
      for iter in tqdm(inter_text_list):
        if(m_name in iter or m_name.lower() in iter or m_name.upper() in iter or m_name.title() in iter ):
          inter_dict["sentence"]=iter
          inter_dict["method"]=m_name
        interd_list.append(inter_dict)          
    doc_level_ann["abstract"]=(interd_list)
  for body in tqdm(item["body_text"]):
    inter_text_list = body["text"].split(".")
    for m_name in tqdm(method_list):
      for iter in tqdm(inter_text_list):
        if(m_name in iter or m_name.lower() in iter or m_name.upper() in iter or m_name.title() in iter ):
          inter_dict["sentence"]=iter
          inter_dict["method"]=m_name
          interd_list.append(inter_dict)
    doc_level_ann["body_text"]=(interd_list)
  doc_level_ann["paper_id"]=item["paper_id"]
  doc_level_list.append(doc_level_ann)
  with open("/content/drive/MyDrive/doc_data.json", 'w') as f:
    json.dump(doc_level_list, f)
  break

```
